### Model Comparation by Accuracy

#### Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
perceptron_accuracy = 0.9824561403508771
adaline_gd_accuracy = 0.6257309941520468
adaline_sgd_accuracy = 0.6257309941520468
adaline_bgd_accuracy = 0.6257309941520468
```

#### Plot the comparison

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(['Perceptron', 'Adaline GD', 'Adaline SGD', 'Adaline BGD'], [perceptron_accuracy, adaline_gd_accuracy, adaline_sgd_accuracy, adaline_bgd_accuracy])
ax.set_ylim(0, 1)
ax.set_ylabel('Accuracy')
ax.set_title('Model Comparison by Accuracy')
plt.show()
```
